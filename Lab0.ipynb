Створити функцію для реалізації алгоритму, відповідно до номеру свого
варіанту з табл. 1.
Реалізація гіпотези Гольдбаха, що стверджує, що будь-
яке парне число (окрім 2) може бути представлене у

вигляді суми двох простих чисел.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
def goldbach_check(number):
    for i in range(2, number // 2 + 1):
        if is_prime(i) and is_prime(number - i):
            return (i, number - i)
        
originalNumber = input("Enter the even number:")
originalNumber = int(originalNumber)
while originalNumber <= 2 or originalNumber % 2 != 0: 
    originalNumber = input("Enter the even number:")
    originalNumber = int(originalNumber)
print(f"Число {originalNumber} можна представити як {goldbach_check(originalNumber)[0]} + {goldbach_check(originalNumber)[1]}")
```
